```python
import pandas as pd
import numpy as np

```

```python

data_train = pd.read_csv('./data/films.csv')
data_train
```

```python
colunms = [
    'year',
    'rating',
    'happy',
    'action',
    'weird'
]
```

```python
X_train = data_train[colunms]
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(n_neighbors=3)
model.fit(X_train)
```

```python
x = np.array([1999, 7, 2, 9, 3]).reshape(1, -1)
distances, idx = model.kneighbors(x)
```

```python
films = []
for i in idx.ravel():
    films.append(data_train.iloc[i]['title'])
```

```python
print(idx, films)
print(distances)
```
